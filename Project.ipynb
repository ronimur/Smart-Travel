```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sum the payoff of each unique player
player_payoff_sum = df.groupby('Player')['payoff'].sum().reset_index()

# Rank the players by their payoff
player_payoff_sum['rank'] = player_payoff_sum['payoff'].rank(method='first', ascending=False)
player_payoff_sum.sort_values(by='rank', inplace=True)

# Calculate the 25th and 75th percentiles of the payoff
payoff_25_quantile = player_payoff_sum['payoff'].quantile(0.25)
payoff_75_quantile = player_payoff_sum['payoff'].quantile(0.75)


# Re-categorize players into above 75% and below 25% quantiles, excluding the middle category
player_payoff_sum['category'] = pd.cut(player_payoff_sum['payoff'],
                                       bins=[-float('inf'), payoff_25_quantile, float('inf')],
                                       labels=['Below 25%', 'Above 75%'])

# Filter data to include only the below 25% and above 75% categories
filtered_payoff_sum = player_payoff_sum[player_payoff_sum['category'].isin(['Below 25%', 'Above 75%'])]

# Plot the box plot comparing below 25% and above 75% payoffs
plt.figure(figsize=(8, 3))
sns.boxplot(x='category', y='payoff', data=filtered_payoff_sum, palette='viridis')
plt.title('Payoff Distribution: Below 25% vs Above 75%')
plt.xlabel('Category')
plt.ylabel('Total Payoff')
plt.grid(True)

plt.show()


```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

features = df.groupby('Player')[[
    'decision_Private Ride', 'decision_Shared Ride', 'decision_Automated Transit', 
     'realtrafficTime', 'time_respond', 'departureTime', 
    'scenario', 'accident', 'gender', 'payoff'
]].mean().reset_index()

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features[['decision_Private Ride', 'decision_Shared Ride', 'decision_Automated Transit', 
     'realtrafficTime', 'time_respond', 'departureTime', 'scenario', 'accident','payoff']])

# Elbow Method
wcss = []
silhouette_scores = []
K = range(2, 11)  # Trying different values of k from 2 to 10

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(scaled_features, kmeans.labels_))

# Plotting the Elbow Method graph
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(K, wcss, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal k')

# Plotting the Silhouette Score graph
plt.subplot(1, 2, 2)
plt.plot(K, silhouette_scores, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal k')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier

# Function to perform K-Means clustering and add cluster labels to the original dataframe
def add_cluster_labels(df_scenario, optimal_k=3):
    features = df_scenario.groupby('Player')[[
        'decision_Private Ride', 'decision_Shared Ride', 'decision_Automated Transit', 
        'payoff', 'realtrafficTime', 'time_respond', 'departureTime', 
        'accident', 'gender'
    ]].mean().reset_index()

    # Standardize the features
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features.drop(columns=['Player']))

    # Fit K-Means with the optimal number of clusters
    kmeans = KMeans(n_clusters=optimal_k, random_state=42)
    features['Cluster'] = kmeans.fit_predict(scaled_features)

    # Merge the cluster labels back to the original dataframe
    df_scenario = df_scenario.merge(features[['Player', 'Cluster']], on='Player')
    return df_scenario, features, scaler, kmeans

# Function to plot feature importances
def plot_feature_importances(features, feature_columns, scaler, kmeans, scenario_number):
    # Standardize the features
    scaled_features = scaler.fit_transform(features[feature_columns])

    # Train a Random Forest classifier to predict cluster labels
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(scaled_features, features['Cluster'])

    # Get feature importances
    feature_importances = rf.feature_importances_

    # Ensure the lengths match
    assert len(feature_columns) == len(feature_importances), "Lengths of feature names and importances must match"
    # Define feature columns (excluding 'Player' and 'Cluster')
    feature_columns = [' Private Ride', ' Shared Ride',  'Automated Transit', 
                     'payoff', 'traffic Time', 'response time', 'departure Time', 'accident', 'gender']
    # Create a DataFrame for visualization
    importances_df = pd.DataFrame({'Feature': feature_columns, 'Importance': feature_importances})
    importances_df = importances_df.sort_values(by='Importance', ascending=False)

    # Plot feature importances
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Importance', y='Feature', data=importances_df, color='#01263D')
    plt.title(f'Feature Importances for Cluster Prediction, Scenario {scenario_number}')
    plt.show()

    # Display feature importances
    print(importances_df)

# Separate the data by scenario
df_scenario1 = df[df['scenario'] == 1]
df_scenario2 = df[df['scenario'] == 2]

# Add cluster labels to each scenario
df_scen1, features_scen1, scaler_scen1, kmeans_scen1 = add_cluster_labels(df_scenario1, 3)
df_scen2, features_scen2, scaler_scen2, kmeans_scen2 = add_cluster_labels(df_scenario2, 3)

# Define feature columns (excluding 'Player' and 'Cluster')
feature_columns = ['decision_Private Ride', 'decision_Shared Ride', 'decision_Automated Transit', 
                   'payoff', 'realtrafficTime', 'time_respond', 'departureTime', 'accident', 'gender']

# Plot feature importances for Scenario 1
plot_feature_importances(features_scen1, feature_columns, scaler_scen1, kmeans_scen1, 1)

# Plot feature importances for Scenario 2
plot_feature_importances(features_scen2, feature_columns, scaler_scen2, kmeans_scen2, 2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Separate the data by scenario
df_scenario1 = df[df['scenario'] == 1]
df_scenario2 = df[df['scenario'] == 2]

# Function to perform K-Means clustering and calculate average payoffs by gender and cluster
def kmeans_clustering_and_payoff_comparison(df_scenario, scenario_number):
    gender_labels = {0: 'Men', 1: 'Women'}
    comparison_results = []

    for gender_value in [0, 1]:
        features = df_scenario[df_scenario['gender'] == gender_value].groupby('Player')[[
            'decision_Private Ride', 'decision_Shared Ride', 'decision_Automated Transit', 
            'payoff', 'realtrafficTime', 'time_respond', 'departureTime', 
            'accident', 'gender'
        ]].mean().reset_index()

        # Standardize the features
        scaler = StandardScaler()
        feature_columns = ['decision_Private Ride', 'decision_Shared Ride', 'decision_Automated Transit', 
                           'payoff', 'realtrafficTime', 'time_respond', 'departureTime', 'accident', 'gender']
        scaled_features = scaler.fit_transform(features[feature_columns])

        # Fit K-Means with the optimal number of clusters
        optimal_k = 3  # Based on previous findings
        kmeans = KMeans(n_clusters=optimal_k, random_state=42)
        features['Cluster'] = kmeans.fit_predict(scaled_features)

        # Calculate average payoff for each cluster
        cluster_avg_payoff = features.groupby('Cluster')['payoff'].mean()
        comparison_results.append((gender_value, cluster_avg_payoff))

        # Print cluster centroids
        centroids = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_), columns=feature_columns)
        print(f"Cluster Centroids (Scenario {scenario_number}, Gender {gender_labels[gender_value]}):\n", centroids)
        print(f"Average Payoff by Cluster (Scenario {scenario_number}, Gender {gender_labels[gender_value]}):\n", cluster_avg_payoff)

    return comparison_results

# Apply the function to each scenario
results_scenario1 = kmeans_clustering_and_payoff_comparison(df_scenario1, 1)
results_scenario2 = kmeans_clustering_and_payoff_comparison(df_scenario2, 2)

# Compare the average payoffs and count the comparisons where women have a higher average payoff than men
higher_payoff_count = 0

def compare_payoffs(results_scenario, scenario_number):
    global higher_payoff_count
    for cluster in range(3):  # We have 3 clusters
        men_payoff = results_scenario[0][1][cluster]
        women_payoff = results_scenario[1][1][cluster]
        print(f"Scenario {scenario_number}, Cluster {cluster} - Men: {men_payoff}, Women: {women_payoff}")
        if women_payoff > men_payoff:
            higher_payoff_count += 1

print(compare_payoffs(results_scenario1, 2))


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

# Group by gender and scenario and calculate the mean and standard deviation of payoffs
gender_scenario_stats = df.groupby(['gender', 'scenario'])['payoff'].agg(['mean', 'std']).reset_index()

# Convert gender to readable labels
gender_labels = {0: 'Women', 1: 'Men'}
gender_scenario_stats['gender'] = gender_scenario_stats['gender'].map(gender_labels)

# Plot the payoff difference by gender and scenario
plt.figure(figsize=(10, 5))
sns.barplot(x='scenario', y='mean', hue='gender', data=gender_scenario_stats)
plt.title('Average Payoff by Gender and Scenario')
plt.xlabel('Scenario')
plt.ylabel('Average Payoff')
plt.legend(title='Gender')
plt.show()

# Perform t-test for each scenario
scenarios = df['scenario'].unique()
for scenario in scenarios:
    men_payoff = df[(df['gender'] == 0) & (df['scenario'] == scenario)]['payoff']
    women_payoff = df[(df['gender'] == 1) & (df['scenario'] == scenario)]['payoff']
    
    t_stat, p_value = ttest_ind(men_payoff, women_payoff, equal_var=False)
    print(f"Scenario {scenario} - T-test Results:")
    print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}\n")

# Display the grouped data with mean and standard deviation
print(gender_scenario_stats)
```
